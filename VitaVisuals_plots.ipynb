```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
training = pd.read_csv('Training.csv')
symptom_data = training.drop(columns=['prognosis'])
symptom_frequencies = symptom_data.sum().sort_values(ascending=False)
plt.figure(figsize=(15, 10))
sns.barplot(x=symptom_frequencies.values, y=symptom_frequencies.index, palette="viridis")
plt.title('Frequency of Symptoms')
plt.xlabel('Frequency')
plt.ylabel('Symptoms')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
training = pd.read_csv('Training.csv')
symptom_data = training.drop(columns=['prognosis'])
corr_matrix = symptom_data.corr()
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
plt.title('Symptom Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
training = pd.read_csv('Training.csv')
prognosis_counts = training['prognosis'].value_counts()
plt.figure(figsize=(12, 8))
sns.barplot(x=prognosis_counts.values, y=prognosis_counts.index, palette="muted")
plt.title('Distribution of Prognosis')
plt.xlabel('Number of Cases')
plt.ylabel('Prognosis')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
training = pd.read_csv('Training.csv')
subset_data = training[['itching', 'skin_rash', 'nodal_skin_eruptions', 'continuous_sneezing', 'prognosis']]
melted_data = pd.melt(subset_data, id_vars=['prognosis'], var_name='Symptom', value_name='Presence')
plt.figure(figsize=(15, 10))
sns.boxplot(x='Symptom', y='Presence', hue='prognosis', data=melted_data)
plt.title('Box Plot of Symptom Severity by Prognosis')
plt.xlabel('Symptom')
plt.ylabel('Presence')
plt.xticks(rotation=45)
plt.show()
```
